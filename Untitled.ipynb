```python
ca_data['description'] = ca_data['description'].fillna(value='None')
```

```python
ca_data['like_percentage'] = (ca_data['likes'] / ca_data['views']) * 100
ca_data['dislike_percentage'] = (ca_data['dislikes']/ca_data['views']) * 100
ca_data.loc[ca_data['likes'] >= ca_data['dislikes'],'more_likes'] = 1
ca_data.loc[ca_data['likes'] < ca_data['dislikes'],'more_likes'] = 0
```

```python
def get_title_length(title):
    return len(title)
def get_number_of_tags(string_of_tags):
    tags = string_of_tags.split('|')
    tags_no_empty_strings = []
    for tag in tags:
        if tag != '':
            tags_no_empty_strings.append(tag)
    return len(tags_no_empty_strings)

ca_data['title_length'] = ca_data.apply(lambda row: get_title_length(row['channel_title']), axis=1)
ca_data['number_of_tags'] = ca_data.apply(lambda row: get_number_of_tags(row['tags']), axis=1)
```

```python
def contains_all_caps_word(title):
    words = title.split(' ')
    for word in words:
        if word.isupper():
            return True 
print(contains_all_caps_word('HEY how are you'))
```

# **Conclusion**

Here are some observations from our exploration that will inform the way we build our model:

1. Movies and Music historically get ma higher number of views than other categories. Including category as a predictor is a good idea
 
2. Likes and Comment count are highly correlated. Since linear regression assumes multicolinearity it could be a good idea to remove one of them
 
3. The standard deviation on month 10 is quite high so it's unreasonable to draw an assumption that there is any statistical significance to videos published in this month getting many views
 
4. Over the few months the trending data was collected, we saw a high amount of views for videos published at 10 AM. There is little standard deviation on this high figure. This relationship with our target variable may mean we can use publish hour as a predictor
 
5. There is a relationship between likes and categories. Does that mean that one must be removed since we're assuming independance?
 
6. The title length and number of tags have a very low correlation with our target variable so it might not be worth including for our model

# CONCLUSION

Our analysis of the YouTube videos dataset has uncovered significant trends and insights into the platform's dynamics. Two key findings stand out prominently:

**Geographic Dominance of US Videos**: A striking observation from our analysis is the overwhelming prevalence of videos originating from the United States. This dominance underscores the immense influence and reach of American content creators on the platform. The abundance of US videos suggests that YouTube's user base is heavily influenced by American culture and content trends, shaping the platform's global landscape.

**Entertainment Reigns Supreme**: Our analysis has also revealed that the entertainment category boasts the highest number of followers among all categories. This finding underscores the enduring popularity of entertainment content on YouTube, indicating that users gravitate towards videos that entertain, amuse, and engage. Whether it's music, movies, gaming, or other forms of entertainment, content creators in this category wield significant influence and attract large audiences.

These insights carry implications for content creators, marketers, and platform developers alike. Understanding the dominance of US videos and the popularity of the entertainment category provides valuable strategic guidance for optimizing content creation, targeting specific audiences, and enhancing user engagement. By leveraging these insights, stakeholders can effectively navigate the competitive landscape of YouTube and capitalize on emerging trends to maximize their impact and reach.

However, it's important to acknowledge that our analysis is based on the available dataset and may not capture the entirety of YouTube's diverse ecosystem. Further research and analysis, including qualitative assessments and user surveys, may be necessary to validate these findings and explore additional dimensions of YouTube's content landscape and user behavior.
